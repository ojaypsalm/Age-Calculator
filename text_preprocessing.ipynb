```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
df = pd.read_csv('/content/tripadvisor_hotel_reviews.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Review'][0]
```

```python
df['review_lowercase'] = df['Review'].str.lower()
```

```python
df['review_lowercase'][0]
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
df['review_lowercase'][0]
```

```python
df['review_no_stopwords']= df['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
df['review_no_stopwords'][0]
```

```python
df['review_no_stopwords_no_punct'] = df.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis =1)
```

```python
df.head()
```

```python
df['review_no_stopwords_no_punct'] = df.apply(lambda x: re.sub(r"([^\w\s])", "", x['review_no_stopwords_no_punct']), axis = 1)
```

```python
df.head()
```

```python
df['tokenized'] = df.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis = 1)
```

```python
df['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
df['stemmed'] = df['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
df['lemmatized'] = df['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
df.head()
```

```python
df['stemmed'][0]
```

```python
df['lemmatized'][0]
```

```python
token_clean = sum(df['lemmatized'], [])
```

```python
unigram =(pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
```

```python
print(unigram)
```

```python
bigram =(pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigram)
```
